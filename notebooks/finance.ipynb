```python
import pandas as pd
import numpy as np

file_path = "Apple (AAPL) From 1980 To Dec-2024.csv"
df = pd.read_csv(file_path)

df.columns = df.columns.str.strip()
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', utc=True)
df = df.dropna(subset=['Date'])
df = df.sort_values('Date')
df.set_index('Date', inplace=True)

prices = df['Close'].astype(float)
returns = prices.pct_change().dropna()

rolling_volatility = returns.rolling(window=30).std() * np.sqrt(252)

confidence_level = 0.95
historical_var = np.percentile(returns, (1 - confidence_level) * 100)

mean_return = returns.mean()
daily_volatility = returns.std()

simulated_returns = np.random.normal(
    mean_return,
    daily_volatility,
    10000
)

monte_carlo_var = np.percentile(simulated_returns, (1 - confidence_level) * 100)

summary = pd.DataFrame({
    "Metric": [
        "Mean Daily Return",
        "Daily Volatility",
        "Historical VaR (95%)",
        "Monte Carlo VaR (95%)"
    ],
    "Value": [
        mean_return,
        daily_volatility,
        historical_var,
        monte_carlo_var
    ]
})

dashboard_data = pd.DataFrame({
    "Price": prices,
    "Daily Return": returns,
    "Rolling Volatility (30D)": rolling_volatility
}).dropna()

mc_days = 252
mc_simulations = 10000

mc_matrix = np.random.normal(
    mean_return,
    daily_volatility,
    (mc_days, mc_simulations)
)

mc_price_paths = prices.iloc[-1] * np.exp(np.cumsum(mc_matrix, axis=0))
mc_df = pd.DataFrame(mc_price_paths)

final_prices = mc_df.iloc[-1]

mc_summary = pd.DataFrame({
    "Metric": [
        "Expected Price",
        "5th Percentile",
        "95th Percentile"
    ],
    "Value": [
        final_prices.mean(),
        np.percentile(final_prices, 5),
        np.percentile(final_prices, 95)
    ]
})

dashboard_data.to_csv("AAPL_AlphaPulse_Dashboard.csv")
summary.to_csv("AAPL_AlphaPulse_Risk_Summary.csv", index=False)
mc_df.to_csv("AAPL_AlphaPulse_MonteCarlo_Paths.csv", index=False)
mc_summary.to_csv("AAPL_AlphaPulse_MonteCarlo_Summary.csv", index=False)

print("AlphaPulse pipeline executed successfully")
```
